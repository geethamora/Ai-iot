```
#step-3:Handling missing data or values
#Senario-1:missing values in numerical column
#caluclate the mean or average of the column
#Senario-2:missing values in non numerical values
#sol-fill the column with high repeated value
#Senario-3:having missing in huge dataset(eg:20000 records in that 2 rows)
#sol:remove those empty values from the dataset
import numpy as np
from sklearn.impute import SimpleImputer
#sklearn:it is a python libaray widely used for machine learning and modelling
#impute:it is part of the sklearn
#simpleImputer:it is class which was import from sklearn.impute
imputer=SimpleImputer(missing_values=np.nan,strategy='mean')
imputer.fit(X[:,1:3])
X[:,1:3]=imputer.transform(X[:,1:3])

```

```
#Step4:handling the catagorial data(instead of number there is text formate is called catagorial data)converting catagorial data into numert from sklearn.preprocessing import OneHotEncoderFfrom sklearn.preprocessing import OneHotEncoder
from  sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer#columntransformer is the class import from sklearn.compose
#it is used to apply changes or transformers to column
#it is class which is used to apply the changes of numerical to text
#Alexa-0,0,1
ct=ColumnTransformer(transformers=[('encoder',OneHotEncoder(),[0])],remainder='passthrough')#encoder is a method which convert text to numeric
X=ct.fit_transform(X)
```
